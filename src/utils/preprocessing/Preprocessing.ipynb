```python
# ==============================================================================================================================

def create_raw(data, ch_names1, ch_names2, sr):
    ch_types = ["eeg" for _ in range(len(ch_names1))]
    info = mne.create_info(ch_names2, ch_types=ch_types, sfreq=sr)
    eeg_data = np.array(data[ch_names1].T, dtype='float') / 1000000 # in Volt # TODO not sure if each dataset is in uv
    raw = mne.io.RawArray(eeg_data, info)
    return raw

# ==============================================================================================================================

from pyprep.prep_pipeline import PrepPipeline

def Union(lst1, lst2):
    final_list = list(set(lst1) | set(lst2))
    return final_list

def find_bad_by_prep(raw, montage_kind, grid_frequency, matlab_strict=False, hacky=False):

    # Add a montage to the data
    montage = mne.channels.make_standard_montage(montage_kind)

    # TODO: change the 50Hz based on location (EU/US)
    prep_params = {
        "ref_chs": "eeg",
        "reref_chs": "eeg",
        "line_freqs": np.arange(grid_frequency, raw.info["sfreq"] / 2, grid_frequency),
    }

    prep = PrepPipeline(
        raw.copy(), prep_params, montage, matlab_strict=matlab_strict
    )

    
    prep.fit()
    raw.info["bads"].extend(
        Union(prep.interpolated_channels, prep.still_noisy_channels)
    )

    # print("BAD CHANNELS: ", Union(prep.interpolated_channels, prep.still_noisy_channels))
    raw.set_montage(montage)

    return raw
```

```python
import os
import numpy as np
import pandas as pd
import re
import sys
from concurrent.futures import ThreadPoolExecutor, as_completed
from tqdm import tqdm
from math import ceil
import json
import time
import mne

import logging

# Set the logging level to ERROR
logging.basicConfig(level=logging.ERROR)
pyprep_logger = logging.getLogger('pyprep')
pyprep_logger.setLevel(logging.ERROR)
numpy_logger = logging.getLogger('numpy')
numpy_logger.setLevel(logging.ERROR)
mne.set_log_level("ERROR")

print("Finished importing")

index_path = '/itet-stor/maxihuber/net_scratch/stuff/cleaning/index_new_6.json'
with open(index_path, 'r') as file:
    index = json.load(file)
print("Index:", len(index))

## ==== store all paths in foundation_prepared will not yet be processed to a file ====

base_dir = '/itet-stor/maxihuber/deepeye_storage/foundation_prepared/'
paths = os.listdir(base_dir)
path_dict = {os.path.join(base_dir, path): False for path in paths}

for index_element in index:
    path_dict[index_element["path"]] = True

missing_paths = sorted([path for path in path_dict.keys() if not path_dict[path]])
print("Skipping", len(missing_paths), "in foundation_prepared")

missing_paths_store_path = '/itet-stor/maxihuber/net_scratch/stuff/cleaning/missing_paths_1.txt'
with open(missing_paths_store_path, 'w') as file:
    for path in missing_paths:
        file.write(path + '\n')

## ==== apply PrepPipeline to elements in index ====

class Preprocessor:
    def __init__(
        self,
        index_path: str,
        index_store_dir: str,
        num_threads: int,
        num_chunks: int
    ):
        self.index_path = index_path
        self.index_store_dir = index_store_dir
        self.num_threads = num_threads
        self.num_chunks = num_chunks

    def preprocess_chunk(self, index_chunk, idx):

        index_chunk_new = []
        
        for i, index_element in enumerate(index_chunk):

            print("=" * 100)
            start_time = time.time()
            
            ## == load == 
            if index_element["path"].endswith(".edf"):
                print("Don't prep-pipeline .edf data")
                bad_channels = None
                
            elif index_element["path"].endswith(".pkl"):
                with open(index_element["path"], "rb") as file:
                    file_data = pd.read_pickle(file)
                # TODO: the region information from the index_element
                raw = create_raw(data=file_data,
                                 ch_names1=index_element["channels"],
                                 ch_names2=index_element["channels_mapped"],
                                 sr=index_element["sr"])
                ## == pre-process .pkl files ==
                try:
                    bad_raw = find_bad_by_prep(raw.copy(), 
                                               montage_kind=index_element["montage"],
                                               grid_frequency=index_element["GridFrequency"])
                    bad_channels = list(bad_raw.info["bads"])
                except Exception as e:
                    print("An error occurred:", e, "for file", index_element['path'])
                    bad_channels = None
    
            ## == Add "PrepPipeline" column (containing bad_channels) ==
            index_element["PrepPipeline"] = bad_channels

            index_chunk_new.append(index_element)
    
            ## == dump index every 1_000 iterations ==
            if i % 1_000 == 0:
                store_path = os.path.join(self.index_store_dir, f'index_with_bad_channels_{idx}.json')
                with open(store_path, 'w') as file:
                    json.dump(index_chunk_new, file, indent=4)
                print(f"Dumped file, idx = {idx}, i = {i}")

            for key, value in index_element.items():
                print(f"{key}:", value)

            print("took", round(time.time() - start_time,2), "seconds")
            
            ## == iterate ==
    
        return f"Completed chunk {idx}"

    def run(self):
        with open(self.index_path, 'r') as file:
            index = json.load(file)
            index = index[:10]

        with tqdm(total=self.num_chunks, desc="Overall Progress") as pbar:
            with ThreadPoolExecutor(max_workers=self.num_threads) as executor:
                futures = []
                
                # split the index file into num_threads many groups
                chunk_size = ceil(len(index) / self.num_chunks)
                chunks = [index[i:i + chunk_size] for i in range(0, len(index), chunk_size)]

                for idx, chunk in enumerate(chunks):
                    futures.append(executor.submit(self.preprocess_chunk, chunk, idx))

                for future in as_completed(futures):
                    result = future.result()
                    print(result)  # Log the completion
                    pbar.update(1)

## ==== here, we could parallelize (num_threads) ====
preprocessor = Preprocessor(index_path='/itet-stor/maxihuber/net_scratch/stuff/cleaning/index_new_6.json',
                            index_store_dir='/itet-stor/maxihuber/net_scratch/stuff/cleaning/',
                           num_threads=1,
                           num_chunks=1)
starting_time = time.time()
preprocessor.run()
print("overall, took", time.time() - starting_time, "seconds.") 
```

```python
import json
import pandas as pd
import numpy as np
from tqdm import tqdm

hw_descriptions_path = '/home/maxihuber/eeg-foundation/indices_and_labels/hardware_paths.json'
with open(hw_descriptions_path, 'r') as file:
    data = json.load(file)

descr_to_paths_path = '/home/maxihuber/eeg-foundation/indices_and_labels/hardware_paths_map.json'
with open(descr_to_paths_path, 'r') as file:
    map_ = json.load(file)

### End-Goal: update this file: /home/maxihuber/eeg-foundation/indices_and_labels/channel_names_new.txt

# for each hardware description, we need to figure out what channels occur, and then we need to map those to .
# we already have some maps in place, but we need to iterate over all files, open them, and make a set of files for each hw description

idx_to_channels = {}
idx_to_channels_mapped = {}
idx_to_channels_path = '/home/maxihuber/eeg-foundation/indices_and_labels/hw_to_channels.json'
idx_to_channels_mapped_path = '/home/maxihuber/eeg-foundation/indices_and_labels/hw_to_channels_prev.json'

for hw_descr_idx in range(0,29):

    ch_names_map_path = f'/home/maxihuber/eeg-foundation/indices_and_labels/tmp/{hw_descr_idx}.json'
    with open(ch_names_map_path, 'r') as file:
        ch_names_map = json.load(file)

    key = list(data.keys())[hw_descr_idx]
    #print("key:", key)

    paths = map_[key]
    
    # iterate over all paths for the given hw description
    # and collect the headers
    channels = set()
    for path in tqdm(paths, desc=f"Processing paths for idx {hw_descr_idx}", leave=True):

        with open(path, 'rb') as file:
            dummy = pd.read_pickle(file)

        channels.update(list(dummy.columns))

    idx_to_channels[hw_descr_idx] = list(channels)
    idx_to_channels_mapped[hw_descr_idx] = list(ch_names_map.values())

    with open(idx_to_channels_path, 'w') as file:
        json.dump(idx_to_channels, file, indent=4)

    with open(idx_to_channels_mapped_path, 'w') as file:
        json.dump(idx_to_channels_mapped, file, indent=4)

print("Done!")
```
